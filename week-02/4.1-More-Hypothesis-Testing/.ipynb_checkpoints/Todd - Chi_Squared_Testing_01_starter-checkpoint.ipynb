Based on 1990 census data, is there a relationship between sex and high-income?

### Chi-Squared Test of Association

This test helps determine if there is a statistically significant correlation between two categories?

$$ \chi^2 = \sum \frac{(observed - expected)^2}{expected} $$

```python
Image(filename="/Users/jam3jam/Desktop/chi.png")
```

The distribution of the $\chi^2$ statistic is chi-square with (r-1)(c-1) degrees of freedom, where r represents the number of rows in the two-way table and c represents the number of columns. The distribution is denoted (df), where df is the number of degrees of freedom.
