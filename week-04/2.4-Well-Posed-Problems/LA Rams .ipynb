Pressure unit in SI system is pascal (derived SI unit not base unit). It is equivalent to one newton per square meter. Other common units are kilopascal (kpa), megapascal (mpa), psi (pound per square inch), torr (mmHg), atm (atmospheric pressure) and bar. Pressure is the amount of force applied per unit area.
