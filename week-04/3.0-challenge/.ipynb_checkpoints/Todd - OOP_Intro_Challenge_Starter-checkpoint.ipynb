## Objects

#### combine state (data) and behavior (algorithms)

#### Objects decide what is exposed to the outside world (by their public interface) and hide their implementation details to provide abstraction.

```python
class Dog():
    name = "roonie"
    type_dog = "poodle"
    bark_protocol = 1
    fleas = False
    fav_command = "fetch"
    bark_counter = 0
    
    def __init__(self, name=name, type_dog = type_dog, fleas=fleas, fav_command=fav_command, bark_protocol = 5, bark_counter = 0):
        self.name = name
        self.type_dog = type_dog
        self.fleas = fleas
        self.fav_command = fav_command
        self.fav_command = fav_command
        self.bark_protocol = bark_protocol

        self.bark_counter = 0
        self.miles_walked = 0
        
    def walk(self):
        self.miles_walked +=1
        
    def bark(self):

        for i in range(self.bark_protocol):
            self.bark_counter += 1
            print "woof, woof"
```

```python
mike = Dog(name="mike")
mike.walk()
```

```python
mike.bark()
```
