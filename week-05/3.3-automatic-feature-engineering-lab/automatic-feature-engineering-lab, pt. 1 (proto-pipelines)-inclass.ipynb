```python
iowa_liquor_store_subset_df = (iowa_liquor_store_df[["zip_code", "sale_2016", "store_count_2016", "sale_2015", "store_count_2015",
                                                            "sale_2014", "store_count_2014", "sale_2013", "store_count_2013",
                                                            "sale_2012", "store_count_2012"]])
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def plot_zips(s):
    df = iowa_liquor_store_subset_df[iowa_liquor_store_subset_df["zip_code"].isin(s)]
    colors = ["b", "g", "r", "c", "m"]
    years = [2012, 2013, 2014, 2015, 2016]
    fig, ax = plt.subplots(figsize=(20, 10))
    n = df.shape[0]
    ind = np.arange(n)  # the x locations for the groups
    width = 0.1       # the width of the bars

    rects = []
    legs = []
    for i, year in enumerate(years):
        col = "sale_" + str(year)
        rects.append(ax.bar(1.5*ind+i*width, df[col], width, color=colors[i]))
        legs.append((year, colors[i]))

        ax.set_xticks(1.5*ind + 2*width)
        ax.set_xticklabels(df["zip_code"], rotation="vertical")
    ax.legend(rects, years)
    plt.show

zipcodes = [50010, 50020, 50320, 50314, 50310, 50321, 50322]
plot_zips(zipcodes)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

def least_squares_regressor_standard_workflow(X, y, random_state=None):
    # Receive the data
    # Split the data into training and testing sets; be sure to include an option to specify a random state
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=random_state)

    # Create a new OLS model 
    model = LinearRegression()
    
    # Fit the model
    model.fit(X_train, y_train)
    
    # Score the model
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    
    # Return a dictionary containing the model, the train score, and the test score
    return {
        'model' : model,
        'train_score' : train_score,
        'test_score' : test_score
    }
```
